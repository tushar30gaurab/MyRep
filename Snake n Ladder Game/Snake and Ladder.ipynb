```python
                        ####################################
                        #                                  #
                        #      SNAKE AND LADDERS GAME       #
                        #                                  #
                        ####################################

import time
import random
import sys


SYSTEM_HALT = 1           # System gets halt for 1 second.
MAXIMUM_VALUE = 100       # Maximum number of blocks in the Game.
DICE_FACE = 20            # Total 'FACES' of a dice ( Here it is 20, as per the problem statement ).



# Snake takes you DOWN from 'key' to 'value'
snakes = {18: 3 , 39: 5 , 51: 6 , 54: 36 , 56: 1 , 60: 23 , 75: 28 , 83: 45 , 85: 59 , 90: 48 , 92: 25 , 97: 87 , 99: 63 }
    



# Ladder takes you UP from 'key' to 'value'
ladders = { 4: 21 , 6: 13 , 11: 27 , 17: 74 , 22: 37 , 38: 59 , 49: 67 , 57: 76 , 61: 78 , 73: 86 , 81: 98 ,  88: 91 }
    
    



def welcome_msg():
    msg = """
                    ####  Welcome to Snake and Ladder Game  ####

                                ## AUTO MODE ##

                        ####################################
                        #                                  #
                        #         SNAKE AND LADDER         #
                        #                                  #
                        ####################################



    Rules:
        #  Initally both the players are at starting position i.e. 0. 
        #  If you lands at the bottom of a ladder, you can move up to the top of the ladder.
        #  If you lands on the head of a snake, you will slide down to the bottom of the snake.
        #  The first player to get to the FINAL position is the winner.
        #  Hit enter to roll the dice.

    """
    print(msg)

def welcome_msg1():
    msg = """
                    ####  Welcome to Snake and Ladder Game  ####

                                ## MANUAL MODE ##


                         ####################################
                         #                                  #
                         #         SNAKE AND LADDER         #
                         #                                  #
                         ####################################



    Rules:
        #  Initally both the players are at starting position i.e. 0. 
        #  If you land at the bottom of a ladder, you can move up to the top of the ladder.
        #  If you land on the head of a snake, you will slide down to the bottom of the snake.
        #  The first player to get to the FINAL position is the winner.
        #  Enter a Number between 1 and 20 to roll the dice.

    """
    print(msg)


def getPlayerNames():
    player1 = None
    while not player1:
        player1 = input("Enter the name of Player 1: ").strip()

    player2 = None
    while not player2:
        player2 = input("Enter the name of Player 2: ").strip()

    print("## Let us Start ##\n")
    return player1, player2




def GetDiceValue():
    time.sleep(SYSTEM_HALT)
    DiceValue = random.randint(1, DICE_FACE)
    print("You got a " + str(DiceValue))
    return DiceValue




def GetDiceValue1():
    time.sleep(SYSTEM_HALT)
    diceValue = int(input("Enter any number from 1 to 20: "))
    return diceValue



def gotSnakeBite(OldValue, CurrentValue, PlayerName):
    print("\n\nALAS !!\n" + PlayerName + " got a snake bite. Slipped down from " + str(OldValue) + " to " + str(CurrentValue))
    print("\nYour final position is: "+ str(CurrentValue))




def gotLadderJump(OldValue, CurrentValue, PlayerName):
    print("\n\nHURRAY !!\n" + PlayerName + " climbed the ladder from " + str(OldValue) + " to " + str(CurrentValue))
    print("\nYour final position is: "+ str(CurrentValue))




def snake_ladder(PlayerName, CurrentValue, DiceValue):
    time.sleep(SYSTEM_HALT)
    OldValue = CurrentValue
    CurrentValue = CurrentValue + DiceValue

    if CurrentValue > MAXIMUM_VALUE:
        print("You need " + str(MAXIMUM_VALUE - OldValue) + " to win this game. Keep trying.")
        return OldValue

    print("\n" + PlayerName + " moved from " + str(OldValue) + " to " + str(CurrentValue))
    print("\nYour final position is: "+ str(CurrentValue))


    if CurrentValue in snakes:
        final_value = snakes.get(CurrentValue)
        gotSnakeBite(CurrentValue, final_value, PlayerName)


    elif CurrentValue in ladders:
        final_value = ladders.get(CurrentValue)
        gotLadderJump(CurrentValue, final_value, PlayerName)


    else:
        final_value = CurrentValue

    return final_value




def check_win(PlayerName, position):
    time.sleep(SYSTEM_HALT)
    if MAXIMUM_VALUE == position:
        print("\n\n\nThats it.\n\n" + PlayerName + " won the game.\n")
        print("Congratulations " + PlayerName +" !!!")
        print("\n## Game SuccessFully Finished! ##")
        sys.exit(1)




def start():
    ch = str(input("Enter MODE of Game:\n Enter 'A' for Auto Mode and Enter 'M' for Manual Mode: "))        # M/m for MANUAL MODE......A/a for AUTO MODE
    if(ch == 'a' or ch == 'A'):
        welcome_msg()
        time.sleep(SYSTEM_HALT)
        player1, player2 = getPlayerNames()
        time.sleep(SYSTEM_HALT)

        currentPositionP1 = 0      #Player 1 initial position.
        currentPositionP2 = 0      #Player 2 initial position.


        while True:
            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND

            input_1 = input("\n" + player1 + ": "  + " Press 'ENTER' to roll dice: ")
            print("\nDice is rolling....")
            DiceValue = GetDiceValue()

            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND
            print(player1 + " moving....")
            currentPositionP1 = snake_ladder(player1, currentPositionP1, DiceValue)
            check_win(player1, currentPositionP1)

            input_2 = input("\n" + player2 + ": " + " Press 'ENTER' to roll dice: ")
            print("\nDice is rolling....")
            DiceValue = GetDiceValue()

            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND

            print(player2 + " moving....")
            currentPositionP2 = snake_ladder(player2, currentPositionP2, DiceValue)
            check_win(player2, currentPositionP2)


    if(ch=='m' or ch=='M'):
        welcome_msg1()
        time.sleep(SYSTEM_HALT)
        player1, player2 = getPlayerNames()
        time.sleep(SYSTEM_HALT)

        currentPositionP1 = 0      #Player 1 initial position.
        currentPositionP2 = 0      #Player 2 initial position.


        while True:
            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND

            print(player1 + "'s Turn: ")
            
            DiceValue = GetDiceValue1()
            if(DiceValue>20):
                print("Numbers must be entered within the range , i,e., between 1 and 20. Enter Again!")
                DiceValue = GetDiceValue1()
            print("\nDice is rolling....")

            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND

            print(player1 + " moving....")
            currentPositionP1 = snake_ladder(player1, currentPositionP1, DiceValue)
            check_win(player1, currentPositionP1)

            print(player2 + "'s Turn: ")
            
            DiceValue = GetDiceValue1()
            if(DiceValue>20):
                print("Numbers must be entered within the range , i,e., between 1 and 20. Enter Again!")
                DiceValue = GetDiceValue1()
            print("\nDice is rolling....")

            time.sleep(SYSTEM_HALT)     #SYSTEM GETS HALT FOR 1 SECOND

            print(player2 + " moving....")
            currentPositionP2 = snake_ladder(player2, currentPositionP2, DiceValue)
            check_win(player2, currentPositionP2)




if __name__ == "__main__":
    start()
```
